**Q4**

Gradient of a function is basically partial derivation with respect to each variable

$ f(x,y) = x^{2}y + y^{3}sin(x) $

$ \frac{\partial f(x, y) }{∂x} = 2xy + y^{3}cos(x)$

$ \frac{\partial f(x, y) }{∂y} = x^{2}+3y^{2}sin(x)$

$ [\partial f/\partial x, \  \partial f/\partial y] = [2xy + y^{3}cos(x), \ x^{2}+3y^{2}sin(x)]$

```python
### Q6

import sympy

x, y = sympy.symbols('x y')
f = x**2 * y + y**3 *   sympy.sin(x)

gradient_x = sympy.diff(f, x)
gradient_y = sympy.diff(f, y)

print(f"{gradient_x}")
print(f"{gradient_y}")
```

```python
class Year:
  def __init__(self, year, branches):
    self.year = year
    self.branches = branches

class Branch:
  def __init__(self, name, students):
    self.name = name
    self.students = students

class Student:
  def __init__(self, roll_num, name, marks):
    self.roll_num = roll_num
    self.name = name
    self.marks = marks

student = Student(1, "N", {"Maths": 100, "English": 70})

branch1 = Branch("Branch 1", [student])
branch2 = Branch("Branch 1", [])
branch3 = Branch("Branch 2", [])

year1 = Year(2022, [branch1, branch2])
year2 = Year(2023, [branch2, branch3])
year3 = Year(2024, [branch2, branch3])
year4 = Year(2025, [branch2, branch3])

database = [year1, year2, year3, year4]
```

**Q9**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0.5, 100.01, 0.5)

y = x
plt.plot(x, y)
plt.title(' y = x')
plt.show()

y = x**2
plt.plot(x, y)
plt.title(' y = x^2')
plt.show()

y = (x**3/100)
plt.plot(x, y)
plt.title(' y = x^3/100')
plt.show()

y = np.sin(x)
plt.plot(x, y)
plt.title(' y = sin(x)')
plt.show()

y = np.sin(x)/x
plt.plot(x, y)
plt.title(' y = sin(x)/x')
plt.show()

y = np.log(x)
plt.plot(x, y)
plt.title(' y = log(x)')
plt.show()

y = np.exp(x)
plt.plot(x, y)
plt.title(' y = e^x')
plt.show()
```
